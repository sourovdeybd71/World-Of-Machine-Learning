**Reshaping Arrays**

Implement image2vector() that takes an input of shape (length, height, 3) and returns a vector of shape (length*height*3, 1). For example, if you would like to reshape an array v of shape (a, b, c) into a vector of shape (a*b,c) you would do:

v = v.reshape((v.shape[0]*v.shape[1], v.shape[2])) # v.shape[0] = a ; v.shape[1] = b ; v.shape[2] = c

```python
def image2vector(image):
  v=image.reshape(image.shape[0]*image.shape[1]*image.shape[2], 1)
  return v
```

```python
# This is a 3 by 3 by 2 array, typically images will be (num_px_x, num_px_y,3) where 3 represents the RGB values
image = np.array([[[ 0.67826139,  0.29380381],
        [ 0.90714982,  0.52835647],
        [ 0.4215251 ,  0.45017551]],

       [[ 0.92814219,  0.96677647],
        [ 0.85304703,  0.52351845],
        [ 0.19981397,  0.27417313]],

       [[ 0.60659855,  0.00533165],
        [ 0.10820313,  0.49978937],
        [ 0.34144279,  0.94630077]]])

print ("image2vector(image) = " + str(image2vector(image)))
```

**VECTORIZATION**

*Classic Methods*

```python
import numpy as np
import time

x1 = [1, 2, 3]
x2 = [4, 5, 6]


#dot product (Classic Method)
tic = time.process_time()
dot=0
for i in range(len(x1)):
  dot+=x1[i]*x2[i]

toc = time.process_time()
print ("dot = " + str(dot) + "\n ----- Computation time = " + str(1000*(toc - tic)) + "ms")

#outer product (Classic Method)
tic = time.process_time()
#create a len(x1)*len(x2) matrix with only zeros
outer = np.zeros((len(x1),len(x2)))
for i in range(len(x1)):
  for j in range(len(x2)):
    outer[i, j] = x1[i]*x2[j]

toc= time.process_time()
print("outer = "+str(outer)+"\n ----- Computation time = " + str(1000*(toc - tic)) + "ms")

#elementwise multiplication (Classic Method)
tic = time.process_time()
mul = np.zeros(len(x1))
for i in range(len(x1)):
    mul[i] = x1[i]*x2[i]

toc = time.process_time()
print ("elementwise multiplication = " + str(mul) + "\n ----- Computation time = " + str(1000*(toc - tic)) + "ms")




```

*Vectorized Methods*

```python
x1 = [1, 2, 3]
x2 = [4, 5, 6]
#dot product
tic = time.process_time()
dot = np.dot(x1,x2)
toc = time.process_time()
print ("dot = " + str(dot) + "\n ----- Computation time = " + str(1000*(toc - tic)) + "ms")

#outer product
tic = time.process_time()
outer = np.outer(x1,x2)
toc = time.process_time()
print ("outer = " + str(outer) + "\n ----- Computation time = " + str(1000*(toc - tic)) + "ms")

#elementwise
tic = time.process_time()
mul = np.multiply(x1,x2)
toc = time.process_time()
print ("elementwise multiplication = " + str(mul) + "\n ----- Computation time = " + str(1000*(toc - tic)) + "ms")
```

**LOSS Calculation**

L1 (y-hat, y) = SUM(i=0 to i=m) |y(i) - y-hat(i)|

L2 (y-hat, y) = SUM(i=0 to i=m) [y(i) - y-hat(i)]^2

```python
def L1(yhat, y):
  loss = sum(abs(yhat-y))
  return loss

yhat = np.array([12, 32, 5])
y = np.array([12, 14, 7])
#12-12=0, 32-14= 18, 5-7=2
#0+18+2=20
#expected loss = 20
print("L1= "+str(L1(yhat, y)))
```

```python
def L2(yhat, y):
  x = yhat - y
  #np.dot(x,x) will do the square calculation
  loss = np.dot(x,x)
  return loss

yhat = np.array([12, 32, 5])
y = np.array([12, 14, 7])

print("L2= "+str(L2(yhat, y)))
```
